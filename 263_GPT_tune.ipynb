## len = 3, with UCLA

```python
def read_text_file(filename):
    array_2d = []
    current_array = []

    with open(filename, 'r') as file:
        for line in file:
            line = line.strip()
            if line:
                current_array.append(line)
            else:
                if current_array:
                    array_2d.append(current_array)
                    current_array = []
        if current_array:
            array_2d.append(current_array)

    return array_2d

filename = "/content/test_question_3_with_ucla.txt"
question_3_with_ucla = read_text_file(filename)
```

```python
conversation_list = []

for question_list in question_3_with_ucla:
  conversation = []
  for question in question_list:
    system_message = [{"role": "user", "content": question }]

    completion = client.chat.completions.create(
      model = "gpt-3.5-turbo",
      messages = conversation + system_message)
    response = str(completion.choices[0].message.content)

    conversation.append({"role": "user", "content": question })
    conversation.append({"role": "assistant", "content": response})
  conversation_list.append(conversation)

conversation_strings = []
for conversation in conversation_list:
  tmp_string = ''
  for line in conversation:
    if line['role'] == 'user':
      tmp_string += f"Question: {line['content']} \n"
    if line['role'] == 'assistant':
      tmp_string += f"Answer: {line['content']} \n \n"
  conversation_strings.append(tmp_string)

df_gpt_3_with_ucla = pd.DataFrame(conversation_strings, columns=['question_answer'])
df_gpt_3_with_ucla['conversation_length'] = 3
df_gpt_3_with_ucla['model'] = 'gpt-3.5-turbo'
```

```python
conversation_list = []

chatbot_persona = {
    "name": "BruinBot",
    "style": "friendly and enthusiastic",
    "background": "knowledgeable about all things UCLA"
}

system_persona = f"As a {chatbot_persona['style']} UCLA chatbot assistant named {chatbot_persona['name']}, you are {chatbot_persona['background']}. You should help the incoming freshmen answer questions about the school."

for question_list in question_3_with_ucla:
  conversation = []
  for question in question_list:
    system_message = [{"role": "user", "content": question }]

    completion = client.chat.completions.create(
      model="ft:gpt-3.5-turbo-0125:personal::9WHOcy38",
      messages = conversation + [{"role": "system", "content": system_persona}] + system_message)
    response = str(completion.choices[0].message.content)

    conversation.append({"role": "user", "content": question })
    conversation.append({"role": "assistant", "content": response})
  conversation_list.append(conversation)

conversation_strings = []
for conversation in conversation_list:
  tmp_string = ''
  for line in conversation:
    if line['role'] == 'user':
      tmp_string += f"Question: {line['content']} \n"
    if line['role'] == 'assistant':
      tmp_string += f"Answer: {line['content']} \n \n"
  conversation_strings.append(tmp_string)

df_tuned_3_with_ucla = pd.DataFrame(conversation_strings, columns=['question_answer'])
df_tuned_3_with_ucla['conversation_length'] = 3
df_tuned_3_with_ucla['model'] = 'personal chatbot'
```
